```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.io.wavfile
from scipy.fftpack import dct
import librosa
from librosa import feature, display
import re
```

```python
#data, fs = read('Audios_keigo/academia1.wav')

s, sr = librosa.load('Audios_keigo/audio4.wav')

s2, sr2 = librosa.load('Audios_keigo/audio5.wav')

plt.plot(s)
plt.title(u'Não há certeza')
plt.show()

plt.plot(s2)
plt.title(u'Há certeza')
plt.show()
```

```python
#Extração de features

mfcc_s = feature.mfcc(s, sr, n_mfcc=13, dct_type=2, norm='ortho')

cent_s = feature.spectral_centroid(s, sr)

plt.figure(figsize=(12, 6))
display.specshow(mfcc_s, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
display.specshow(cent_s, x_axis='time')
plt.colorbar()
plt.title('Centroide')
plt.tight_layout()
plt.show()
```
